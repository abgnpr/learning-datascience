## Distribution Plots

### Distribution Plot

```python
sns.displot(tips_df['tip'], bins=5)
```

```python
sns.displot(crash_df['alcohol'], kde=True)
```

### Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='kde')
```

```python
sns.jointplot(x='alcohol', y='speeding', data=crash_df, kind='hex')
```

### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Matrix Plots

### Heatmaps

```python
sns.heatmap(crash_df[crash_df.columns[:-1]].corr(), annot=True, cmap='Blues')
```

```python
flights_pivtab = flights_df.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights_pivtab, cmap='Blues', linewidth=1)
```

### Cluster Map

### Pair Grid

```python
g = sns.PairGrid(penguins_df, diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(penguins_df, diag_sharey=False, corner=True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

```python
iris_df.head()
```

```python
g = sns.PairGrid(iris_df, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.add_legend()
```

```python
g = sns.PairGrid(iris_df, hue='species', 
                 x_vars=['sepal_length', 'sepal_width'],
                 y_vars=['petal_length', 'petal_width'])
g.map(sns.scatterplot)
g.add_legend()
```

### Facet Grid

```python
fg = sns.FacetGrid(tips_df, col='time', row='smoker')
fg.map(sns.histplot, 'total_bill', bins=8)
```

```python
fg = sns.FacetGrid(tips_df, col='time', row='smoker')
fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.618)
fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.618,
                  col_order=['Dinner', 'Lunch'], palette='Set1')
fg.map(sns.scatterplot, 'total_bill', 'tip', edgecolor='w')
```

```python
kws=dict(s=50, linewidth=0.5, edgecolor='w')
fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', 
                   height=4, aspect=1.3, 
                   hue_order=['Yes', 'No'],
                   hue_kws=dict(marker=['^', 'v']))
fg.map(sns.scatterplot, 'total_bill', 'tip', **kws)
```

```python
fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
fg.map(sns.lineplot, 'solutions', 'score', marker='.')
```

### Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, 
           markers=['o', '^'], 
           scatter_kws={'s': 100, 'linewidths': 0.5, 'edgecolor': 'w'})
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', data=tips_df , height=8, aspect=0.6)
```
