# Univariate Analysis

Imports

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.info()
```

### Histogram

```python
plt.figure(figsize=(16, 10))
fig = sns.histplot(data=df, x=df['bill_length_mm'])
fig.set_title('Univariate analysis of Bill Length', fontsize=20, pad=20)
fig.set_xlabel('Bill Length (mm)', fontsize=16, labelpad=20)
fig.set_ylabel('Record Count',  fontsize=16, labelpad=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.draw()
```

### Boxplot

```python
plt.figure(figsize=(16, 10))
fig = sns.boxplot(data=df, x=df['body_mass_g'])
fig.set_title('Univariate Analysis of Body Mass of Penguins', fontsize=20, pad=20)
fig.set_xlabel('Body Mass (g)', fontsize=16, labelpad=20)
plt.xticks(fontsize=14)
plt.draw()
```

### Violin Plot

```python
plt.figure(figsize=(16, 10))
fig = sns.violinplot(data=df, x=df['body_mass_g'])
fig.set_title('Univariate Analysis of Body Mass of Penguins', fontsize=20, pad=20)
fig.set_xlabel('Body Mass (g)', fontsize=16, labelpad=20)
plt.xticks(fontsize=16)
plt.draw()
```

### Bar Plot

```python
df['species'].value_counts()
```

```python
plt.figure(figsize=(16, 10))
fig = sns.countplot(data=df, x=df['species'], hue='species')
fig.set_title('Univariate Analaysis of Pengunin Species', fontsize=20, pad=20)
fig.set_xlabel('Species', fontsize=18, labelpad=20)
fig.set_ylabel('Count', fontsize=18, labelpad=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.draw()
```

```python
plt.figure(figsize=(16, 10))
penguins_group = df.groupby('species').count()
penguins_group.reset_index(inplace=True)
plt.pie(penguins_group['body_mass_g'],
        labels=penguins_group['species'],
        colors=['g', 'b', 'r'],
        textprops={'fontsize': 18})
plt.title('Univariate Analysis of Body Weight of Penguins', fontsize=20)
plt.draw()
```
