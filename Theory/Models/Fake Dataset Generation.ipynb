```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn.datasets as datasets
```

Checkerboard

```python
data, rows, cols = datasets.make_checkerboard(
    shape=(300, 300),
    n_clusters=(4, 4),
    # n_clusters=1,
    shuffle=False,
    noise=5,
    random_state=67
)

print(data)

plt.matshow(data, cmap=plt.cm.Blues)
plt.draw()
```

Classification

```python
X, y = datasets.make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=4,
    n_redundant=1,
    n_repeated=0,
    n_clusters_per_class=1,
    shuffle=False,
    random_state=0
)

frame = pd.DataFrame(np.hstack([X, y.reshape(-1, 1)]))
# frame = np.sort(frame)
print(frame)

plt.figure(figsize=(15,15))

ax = sns.heatmap(frame.iloc[:, :10], cmap='Blues')
# sns.heatmap(frame[[10]], cmap='Reds', ax=ax)

# fig, ax = plt.subplots()
# ax = sns.heatmap(df[[column]], cmap='')
# sns.heatmap(X, cbar=False, xticklabels=False, yticklabels=False, cmap={})

plt.draw()
```
